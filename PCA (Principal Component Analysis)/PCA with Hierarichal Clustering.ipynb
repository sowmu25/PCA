```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df =pd.read_csv(r"wine.csv")
```

```python
df
```

```python
def norm_data(i):
    x= (i - i.min())/(i.min() -i.max())
    return x
```

```python
normalized_data =norm_data(df)
```

```python
normalized_data
```

Applying the Hierarchal Clustering on Normalized data

```python
from sklearn.cluster import  AgglomerativeClustering
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(normalized_data,method ="complete"))
```

```python
clusters =AgglomerativeClustering(n_clusters =3, affinity='euclidean',linkage="ward")
clustering =clusters.fit(normalized_data)
clustering.labels_
```

```python
df['BeforePCA']=clustering.labels_
```

```python
df
```

Applying The Prncipal Component Analysis


```python
from sklearn.decomposition  import PCA
```

```python
pca =PCA(n_components =3,random_state =None)
pca_val =pca.fit_transform(normalized_data)
pca_val
```

```python
final_df =pd.DataFrame(data =pca_val,columns=["Pca1","Pca2","Pca3"])
final_df
```

```python
pca.explained_variance_ratio_
```

```python
dff =pd.DataFrame(pca.explained_variance_ratio_,columns= ["pcas"])
dff1=pd.DataFrame(np.cumsum(np.round(dff['pcas'],decimals =4)*100))
dff
```

```python
dff1
```

```python
#Plotting the curve
plt.plot(dff1,color="green")
```
